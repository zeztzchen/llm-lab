## Lp norm

$$
\|x\|_p=\left(\sum_i^nx_i^p\right)^{1/p}
$$

- $L_p$ norm
    - $L_1$: $\|x\|_1=\sum_i|x_i|$
    - $L_2$: $\|x\|_2=(\sum_ix_i^2)^{1/2}$
    - $L_{\infty}$: $\|x\|_\infty=\max_i|x_i|$


### 可视化、几何分析

$$
\|x\|_p=1
$$

- 表示的是一个集合
- 这个集合在二维空间中是一个闭合曲线，在三维空间中是一个闭合曲面
- 下面使用一个二个值的向量来可视化不同 $p$ 值对应的集合形状

```python
import numpy as np
import matplotlib.pyplot as plt

def lp_unit_circle(p, num_points=1000):
    theta = np.linspace(0, 2 * np.pi, num_points)

    if p == np.inf:
        # L∞ unit ball: max(|x1|, |x2|) = 1
        x = np.array([-1, 1, 1, -1, -1])
        y = np.array([-1, -1, 1, 1, -1])
        return x, y

    x = np.sign(np.cos(theta)) * np.abs(np.cos(theta)) ** (2 / p)
    y = np.sign(np.sin(theta)) * np.abs(np.sin(theta)) ** (2 / p)

    return x, y


ps = [1, 2, 3, np.inf]
labels = ["L1 norm", "L2 norm", "L3 norm", "L∞ norm"]

plt.figure(figsize=(7, 7))

for p, label in zip(ps, labels):
    x, y = lp_unit_circle(p)
    plt.plot(x, y, label=label, linewidth=2)

plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)

plt.gca().set_aspect("equal", adjustable="box")
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("L1, L2, L3, and L∞ Norm Unit Balls")
plt.legend()
plt.grid(True)

plt.show()
```

## matrix norm

- 从向量的 $L_p$ norm 诱导而来（induced norm）
    - biggest $L_p$ norm
    - $\|AB\|_p\leq \|A\|_p\|B\|_p$
    - $\|A\|_1$ 列和最大；
    - $\|A\|_\infty$ 行和最大；
    - $\|A\|_2$ 谱范数；
    - $\|A\|_{\rm F}$ , Frobenius norm.
$$
\begin{split}
&\|A\|_p=\max_{\|x\|_p}\frac{\|Ax\|_p}{\|x\|_p}\\
&\|A\|_1=\max_{1 \leq j \leq n} \sum_{i=1}^m | a_{ij}|\\
&\|A\|_\infty = \max_{1 \leq i \leq m} \sum _{j=1}^n | a_{ij} |\\
&\|A\| _2 = \sigma_{\max}(A) \leq \|A\|_{\rm F} = \left(\sum_{i=1}^m \sum_{j=1}^n |a_{ij}|^2\right)^{\frac{1}{2}} = \left(\sum_{k=1}^{\min(m,n)} \sigma_{k}^2\right)^{\frac{1}{2}}
\end{split}
$$
